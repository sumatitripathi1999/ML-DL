```python
#importing all relavent packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,mean_squared_error, classification_report, r2_score,confusion_matrix,recall_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
#loading the dataset
train=pd.read_csv("/content/train.csv")
test=pd.read_csv("/content/test.csv")
```

```python
train.head(3)
```

```python
#checking for null values------

train.isna().sum()
```

```python
train.dtypes
```

```python
#handling the null values----

# Fill missing values with median for numerical and mode for categorical
numerical_cols=train.select_dtypes(include=[np.number]).columns
train[numerical_cols]=train[numerical_cols].fillna(train[numerical_cols].median())

categorical_cols=train.select_dtypes(include=['object']).columns
train[categorical_cols]=train[categorical_cols].fillna(train[categorical_cols].mode())

# Encode categorical features
for col in train.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])



#assigning our X and Y values--------features and target--

X=train.drop("SalePrice", axis=1)
Y=train['SalePrice']

#spliting the data into training and validation sets------
x_train, x_val, y_train, y_val=train_test_split(X,Y,test_size=0.2,random_state=42)

#standardize the data-----
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_val_scaled=sc.transform(x_val)


```

```python
#Designing the ANN model---------
model=Sequential()

#input layer
model.add(Dense(128,activation='relu',input_shape=(x_train_scaled.shape[1],)))
model.add(Dropout(0.3))

#hidden layer---
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(32,activation='relu'))

#output layer
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python

history = model.fit(x_train_scaled, y_train, validation_data=(x_val_scaled, y_val), epochs=100, batch_size=32)
```

```python
#evaluate the model---------

val_loss=model.evaluate(x_val_scaled,y_val)
print(f"validation_loss={val_loss}")
```

```python
#prediction--------

y_pred=model.predict(x_val_scaled)
```

```python
y_pred
```

```python
rmse=np.sqrt(mean_squared_error(y_val,y_pred))
rmse
```

```python
avg_price = np.mean(y_val)  # Get the average house price
percentage_error = (rmse / avg_price) * 100
print(f'RMSE as a percentage of the average house price: {percentage_error:.2f}%')
```
